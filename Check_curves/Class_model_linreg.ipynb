```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# load and prepare data

df_11_SSP1 = pd.read_csv('C:/Users/toonv/Documents/PBL/Data/linear paths/linear_SSP1_11_2040.csv')
df_11_SSP1_test = pd.read_csv('C:/Users/toonv/Documents/PBL/Data/linear paths/linear_SSP1_11_2050.csv')

df_11_SSP1_test.tail()
```

```python
"""
This class will instatiate a model which finds the reduction that comes with a certain 
carbon price path. 
"""


# model adhv kaj's input

class model:
    
    def __init__(self, num_input):
        
        self.num_inputs = num_input
        self.path = []
        self.reduction = []
        
    def train(self, price_path, reduction):
        
        # input_params is een kostenpad
        # output is de reductie
            
        self.path.append(price_path)
        
        self.path = np.asarray(self.path)
                
        self.reduction.append(reduction)
        
        self.reduction = np.asarray(self.reduction)
        
        df_combined = pd.concat([price_path.reset_index(drop=True),
                                 reduction.reset_index(drop=True)], axis=1)   
        
        return df_combined
            
    def run(self, random_path, df):

        # input params is random costspath    
        self.check = random_path
        
        # when path is included, for now use linear approximation
        y_random_path = self.check
        
        steps = int(10/len(self.check))
        
        x_random_path = np.array(list(range(0, 10, steps))).reshape(-1, 1)                             
                             
        lin_model = LinearRegression()
        
        lin_model.fit(x_random_path, y_random_path)
        
        print('slope random path: ', lin_model.coef_)
        
        key_val = lin_model.coef_ * x_random_path[-1]
        
        # key_val can't be higher than 4000 USD
        if key_val > 4000:
            
            key_val = 4000
        
        print(key_val)
        
        # quick visualisation 
        plt.scatter(x_random_path, y_random_path,  color='black')
        plt.plot(x_random_path, lin_model.coef_*x_random_path, color='blue', linewidth=3)
        plt.xlabel('years')
        plt.ylabel('price')
        
        
        # only use last value of random path
#         key_val = self.check[-1]
        
        self.key_val = key_val
                    
        # find nearest value to key_val
        def find_nearest(array, value):
            
            array = np.asarray(array)
                        
            idx = (np.abs(array - value)).argmin()
            
            return array[idx]
        
        # get last column of path input (for now)
        last_val = []
        
        for path in self.path:
        
            for el in path:
                            
                last_val.append(el[-1])
                                        
        # find values nearest to key_val
        self.nearest1 = find_nearest(last_val, key_val)
        
        print(self.nearest1)
        
        if self.nearest1 == key_val:
            
            index = last_val.index(self.nearest1)
            
            self.nearest2 = last_val[index]
            
            smaller = False
            
            larger = False
        
        elif self.nearest1 < key_val:
            
            # next value is smaller, find value
            index = last_val.index(self.nearest1) + 1
            
            self.nearest2 = last_val[index]
            
            print('smaller', self.nearest1, self.nearest2, key_val)
            
            smaller = True
            
            larger = False
            
        else:
            
            # next value is larger, find value
            index = last_val.index(self.nearest1) - 1
                        
            self.nearest2 = last_val[index]
            
            print('larger', self.nearest1, self.nearest2, key_val)
            
            larger = True
            
            smaller = False
        
        # find reduction corresponding to nearest values in df       
        last_column = df_combined.columns[-2]
                            
        reduction1 = df.loc[df[last_column] == self.nearest1]
        
        reduction2 = df.loc[df[last_column] == self.nearest2]
        
        self.reduction1 = reduction1['reduction']
        
        self.reduction1 = self.reduction1.values[0]
        
        self.reduction2 = reduction2['reduction']
                
        self.reduction2 = self.reduction2.values[0]
        # now that we have x and y, we can do a linear interpolation between the tw
        
        if smaller == True:
            
            diffx = self.nearest1 - self.nearest2
            
            diffy = self.reduction1 - self.reduction2
            
            a = diffy/diffx
                                    
            self.reduction = a * (key_val - self.nearest1)
            
            # y = ax + b, b is self.reduction1 (laagste waarde)
            self.reduction = self.reduction + self.reduction1
                        
        elif larger == True:
            
            diffx = self.nearest2 - self.nearest1
            
            diffy = self.reduction2 - self.reduction1
            
            a = diffy/diffx
                    
            self.reduction = a * (key_val - self.nearest2)
            
            # y = ax + b, b is self.reduction2 (laagste waarde)
            self.reduction = self.reduction + self.reduction2
                                    
        else:
            
            reduction = df.loc[df[last_column] == key_val]
            
            self.reduction = reduction['reduction']

# check if model works
paths = df_11_SSP1.drop(['USD','reduction'], axis = 1)
reduction = df_11_SSP1['reduction']

x = model(5)

df_combined = x.train(paths, reduction)

random_path = sorted(np.random.randint(0, high=3980, size=10))

x.run(random_path, df_combined)

fig, ax = plt.subplots()

print(random_path)

print(x.nearest1, x.nearest2, x.reduction1, x.reduction2)

ax.scatter([x.nearest1, x.nearest2], [x.reduction1, x.reduction2])
ax.scatter(x.key_val, x.reduction, color = 'red')
ax.plot([x.nearest1, x.nearest2], [x.reduction1, x.reduction2], '--')

ax.set_xlabel('price')
ax.set_ylabel('reduction')

x.reduction
```

```python
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

costs = df_11_SSP1['20']
reduction = df_11_SSP1['reduction'].values

print(costs)

# Use only one feature
# costs = costs[:, np.newaxis, 2]

costs_list = []

for cost in costs:
    costs_list.append([cost])
    
costs = costs_list

print(costs[-10:])

# Split the data into training/testing sets, kan niet want pad loopt omhoog
costs_train = costs
costs_test = costs

# Split the targets into training/testing sets
reduction_train = reduction
reduction_test = reduction

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(costs_train, reduction_train)

# Make predictions using the testing set
reduction_pred = regr.predict(costs_test)

# print(reduction_test - reduction_pred)

# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print('Mean squared error: %.2f'
      % mean_squared_error(reduction_test, reduction_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'
      % r2_score(reduction_test, reduction_pred))

# Plot outputs
plt.scatter(costs_test, reduction_test,  color='black')
plt.plot(costs_test, reduction_pred, color='blue', linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```

```python
"""
Linear regression using sklearn, here we try to find the linear fit corresponding to the costcurve.
"""

X = df_11_SSP1.drop('reduction', axis=1)
y = df_11_SSP1['reduction']

X_test = df_11_SSP1_test.drop('reduction', axis=1)
y_test = df_11_SSP1_test['reduction']

linear_regression = linear_model.LinearRegression()

linear_regression.fit(X,y)

y_pred = linear_regression.predict(X)

x_axes = df_11_SSP1.index.values

plt.plot(y_pred, x_axes, color='blue', linewidth=3)
plt.plot(y, x_axes, color='red',linewidth=3)
```

```python
# Ridge regression

"""
Ridge regression addresses some of the problems of Ordinary Least Squares by imposing
a penalty on the size of the coefficients. The ridge coefficients minimize a penalized 
residual sum of squares
"""

ridge_reg = linear_model.Ridge(alpha=.5)

ridge_reg.fit(X,y)

y_pred_ridge = ridge_reg.predict(X)

plt.plot(y_pred_ridge, x_axes, color='blue', linewidth=3)
plt.plot(y, x_axes, color='red',linewidth=3)

score = ridge_reg.score(X,y)

print(score)
```

```python
# Kernel ridge

"""
combines Ridge regression and classification (linear least squares with l2-norm regularization) with the kernel trick.
It thus learns a linear function in the space induced by the respective kernel and the data. For non-linear kernels,
this corresponds to a non-linear function in the original space.
"""

from sklearn.kernel_ridge import KernelRidge

clf = KernelRidge(alpha=1.0)
clf.fit(X, y)

y_pred_clf = clf.predict(X_test)

plt.plot(y_pred_clf, x_axes, color='blue', linewidth=3)
plt.plot(y, x_axes, color='red',linewidth=3)

score = clf.score(X,y)

print(score)
```

```python
from time import time
import numpy as np
import matplotlib.pyplot as plt

from sklearn import linear_model
from sklearn import datasets
from sklearn.svm import l1_min_c

iris = datasets.load_iris()
X = iris.data # dit zou dan de paden van de reductie kunnen zijn
y = iris.target 

X = X[y != 2]
y = y[y != 2]

X /= X.max()  # Normalize X to speed-up convergence

# #############################################################################
# Demo path functions

cs = l1_min_c(X, y, loss='log') * np.logspace(0, 7, 16)


print("Computing regularization path ...")
start = time()
clf = linear_model.LogisticRegression(penalty='l1', solver='liblinear',
                                      tol=1e-6, max_iter=int(1e6),
                                      warm_start=True,
                                      intercept_scaling=10000.)
coefs_ = []
for c in cs:
    clf.set_params(C=c)
    clf.fit(X, y)
    coefs_.append(clf.coef_.ravel().copy())
print("This took %0.3fs" % (time() - start))

coefs_ = np.array(coefs_)
plt.plot(np.log10(cs), coefs_, marker='o')
ymin, ymax = plt.ylim()
plt.xlabel('log(C)')
plt.ylabel('Coefficients')
plt.title('Logistic Regression Path')
plt.axis('tight')
plt.show()
```
